```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import torch
from torch import nn, optim
# from jcopdl.callback import Callback, set_config
from jcopdl.callback import Callback, set_config


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# Dataset & Dataloader

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
bs = 128
crop_size = 64

train_transform = transforms.Compose([
    transforms.RandomRotation(15),
    transforms.RandomResizedCrop(crop_size, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()
])

test_transform = transforms.Compose([
    transforms.Resize(70),
    transforms.CenterCrop(crop_size),
    transforms.ToTensor()
])

train_set = datasets.ImageFolder("data/train/", transform=train_transform)
trainloader = DataLoader(train_set, batch_size=bs, shuffle=True, num_workers=4)

test_set = datasets.ImageFolder("data/test/", transform=test_transform)
testloader = DataLoader(test_set, batch_size=bs, shuffle=True)
```

```python
feature, target = next(iter(trainloader))
feature.shape
```

```python
label2cat = train_set.classes
label2cat
```

# Arsitektur dan Config

```python
# manual
# nn.Conv2d(3, 8, 3, 1, 1),
# nn.ReLU(),
# nn.MaxPool2d(2, 2)
```

```python
from jcopdl.layers import conv_block, linear_block
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            conv_block(3, 8),
            conv_block(8, 16),
            conv_block(16, 32),
            conv_block(32, 64),
            nn.Flatten()
        )
        
        self.fc = nn.Sequential(
            linear_block(1024, 256, dropout=0.1),
            linear_block(256, 2, activation="lsoftmax"),   
        )
        
    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x
```

```python
config = set_config({
    "batch_size": bs,
    "crop_size": crop_size
    })
```

# Training Preparation => MCOC

```python
from model import CNN
```

```python
model = CNN().to(device)
criterion = nn.NLLLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001)
callback = Callback(model, config, outdir="model")
```

# Training

```python
from tqdm import tqdm

def loop_fn(mode, dataset, dataloader, model, criterion, optimizer, device):
    if mode == "train":
        model.train()
    elif mode == "test":
        model.eval()
    cost = correct = 0
    for feature, target in tqdm(dataloader, desc=mode.title()):
        feature, target = feature.to(device), target.to(device)
        output = model(feature)
        loss = criterion(output, target)
        
        if mode == "train":
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        
        cost += loss.item() * feature.shape[0]
        correct += (output.argmax(1) == target).sum().item()
    cost = cost / len(dataset)
    acc = correct / len(dataset)
    return cost, acc
```

```python
while True:
    train_cost, train_score = loop_fn("train", train_set, trainloader, model, criterion, optimizer, device)
    with torch.no_grad():
        test_cost, test_score = loop_fn("test", test_set, testloader, model, criterion, optimizer, device)
    
    # Logging
    callback.log(train_cost, test_cost, train_score, test_score)
#     callback.log("train_cost", train_cost)
#     callback.log("test_cost", test_cost)
#     callback.log("train_score", train_score)
#     callback.log("test_score", test_score)

    # Checkpoint
    callback.save_checkpoint()
        
    # Runtime Plotting
    callback.cost_runtime_plotting()
    callback.score_runtime_plotting()
    
#     callback.add_plot(["train_cost", "test_cost"], scale="semilogy")
#     callback.add_plot(["train_score", "test_score"], scale="linear")
    
    # Early Stopping
    if callback.early_stopping(model, monitor="test_score"):
#         break
        callback.plot_cost()
        callback.plot_score()
        break
```

# Predict

```python
feature, target = next(iter(testloader))
feature, target = feature.to(device), target.to(device)
```

```python
with torch.no_grad():
    model.eval()
    output = model(feature)
    preds = output.argmax(1)
    
preds
```

```python
fig, axes = plt.subplots(6, 6, figsize=(24,24))
for img, label, pred, ax in zip(feature, target, preds, axes.flatten()):
    ax.imshow(img.permute(1, 2, 0).cpu())
    font = {"color": 'r'} if label != pred else {"color": 'g'}
    label, pred = label2cat[label.item()], label2cat[pred.item()]
    ax.set_title(f"label: {label} | pred: {pred} ", fontdict=font);
    ax.axis('off');
```
