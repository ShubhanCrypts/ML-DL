```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

```python
X
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y)
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y1)
ax.plot(X, y2)
```

```python
plt.figure(figsize= (6, 6))

plt.plot(X, y1, 'r-.')
plt.plot(X, y2, 'bv-.')
plt.ylim(-2, 2)
```

```python
plt.figure(figsize= (6, 6))

plt.plot(X, y1, 'r--', linewidth= 0.5)
plt.plot(X, y2, 'bo', markersize=3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title("sin dan cos")
plt.xlabel("X")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r--', linewidth= 0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.axis('equal')
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r--', linewidth= 0.5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2))
ax.set_title("sin dan cos")
ax.set_xlabel("X")
ax.set_ylabel("y")
ax.axis('equal');
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r--', linewidth= 0.5, label="Sin")
ax.plot(X, y2, 'bo', markersize=3, label="Cos")
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel="X", ylabel="y")
ax.axis('equal')
ax.legend(ncol=2, shadow=True, borderpad=1)
```
