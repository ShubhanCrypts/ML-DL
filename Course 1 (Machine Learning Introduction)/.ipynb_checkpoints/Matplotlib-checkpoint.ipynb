```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

```python
X
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y)
```

```python
fig = plt.figure(figsize= (6, 6))
ax = plt.axes()

ax.plot(X, y1)
ax.plot(X, y2)
```
