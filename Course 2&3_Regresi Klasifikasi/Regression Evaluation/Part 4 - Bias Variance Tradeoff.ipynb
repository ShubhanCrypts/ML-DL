### The principle of bias-variance tradeoff

More Complex Model -> low bias, but high variance -> prone to Overfit<br>
More Simple Model -> low variance, but high bias -> prone to Underfit

`increase n_neighbor` => `semakin banyak yang divoting` => `semakin merata` => `complexity decrease` => `prone to Underfit`

Artinya, jika kita training menggunakan KNN dan overfit, maka tambahkan n_neighbor nya

Kernel rbf => lebih cenderung overfit<br>
kernel linear => lebih cenderung underfit

`increase C` => `semakin tidak toleran` => `semakin berliku-liku` => `complexity increase` => `prone to Overfit`

`increase gamma` => `semakin detail` => `semakin berliku-liku` => `complexity increase` => `prone to Overfit`

# Exercise - Random Forest Parameter

- increase n_estimators => more averaging => decrease overfit
- increase max_depth => more branch => low bias => prone to overfit
- increase min_samples_split => prevent splitting => less depth => decrease overfit
- increase max_features => prone to get best decision => prone to overfit
- increase min_impurity_decrease => prevent splitting => decrease overfit
- increase min_samples_leaf => prevent splitting => decrease overfit
