```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Contoh data (X, y)
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Rentang nilai untuk w dan b
w_range = np.linspace(-2, 4, 100)
b_range = np.linspace(-2, 4, 100)

# Membuat grid untuk w dan b
W, B = np.meshgrid(w_range, b_range)

# Fungsi untuk menghitung MSE
def compute_mse(w, b, X, y):
    y_pred = w * X + b
    return np.mean((y - y_pred) ** 2)

# Menghitung MSE untuk setiap kombinasi w dan b
Z = np.array([[compute_mse(w, b, X, y) for w in w_range] for b in b_range])

# Gradient Descent
w = np.random.randn()
b = np.random.randn()
learning_rate = 0.01
n_iterations = 100
history = []

for i in range(n_iterations):
    y_pred = w * X + b
    loss = np.mean((y - y_pred) ** 2)
    history.append((w, b, loss))
    
    dw = -2 * np.mean(X * (y - y_pred))
    db = -2 * np.mean(y - y_pred)
    
    w -= learning_rate * dw
    b -= learning_rate * db

# Membuat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Z, cmap='viridis', alpha=0.6)

# Plot jalur gradient descent
history = np.array(history)
ax.plot(history[:, 0], history[:, 1], history[:, 2], 'r.-', markersize=5, label='Gradient Descent Path')

ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Loss (MSE)')
ax.set_title('Loss Plane and Gradient Descent Path')
ax.legend()

plt.show()
```
